```python
col_preliminary = ['model', 'validation_mse_score', 'test_mse_score', 'training_mse_score', ]
```

```python
preliminary = '/content/drive/My Drive/Pesquisa/PPGIA/gw_ultimate-main/phase2/preliminary/'
```

```python
letters_preliminary = ['A', 'B']
dfs_preliminary = [pd.read_parquet(preliminary + l +'/exp_' + l + '_stats.parquet') for l in letters_preliminary]

df_preliminary = pd.concat(dfs_preliminary)
```

```python
last_epochs = df_preliminary.loc[df_preliminary['epochs'] == '1000']
```

```python
best_preliminary_test = last_epochs[col_preliminary].groupby('model').mean().sort_values('test_mse_score').head(2)
best_preliminary_validation = last_epochs[col_preliminary].groupby('model').mean().sort_values('validation_mse_score').head(2)
best_preliminary_training = last_epochs[col_preliminary].groupby('model').mean().sort_values('training_mse_score').head(2)
```
